```python
client = WatsonMachineLearningAPIClient(wml_credentials)
```

```python
metadata = {
    client.repository.ModelMetaNames.NAME: "Diabetic Retinopathy CNN",
    client.repository.ModelMetaNames.RUNTIME_UID: "tensorflow_1.14-py3.6",
    client.repository.ModelMetaNames.TYPE: "tensorflow_1.14"
    }
model_artifact = client.repository.store_model(model="/project_data/data_asset/neunets_output.tar.gz", 
                                              meta_props=metadata)
```
